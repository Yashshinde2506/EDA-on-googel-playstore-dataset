```python
# replace
df_copy['Size'] = df_copy['Size'].str.replace("M", "000")
```

```python
df_copy['Size'] = df_copy['Size'].str.replace("k","")
```

```python
import numpy as np
```

```python
df_copy['Size'] = df_copy['Size'].str.replace("Varies with device", str(np.nan))
```

```python
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df_copy['Installs']
```

```python
df_copy['Installs'] = df_copy['Installs'].str.replace("+","").str.replace(",","")
```

```python
df_copy['Installs']
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy['Installs']
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Price'] = df_copy['Price'].replace("$","")
df_copy['Price'] = df_copy['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_copy['Price'] = df_copy['Price'].astype(float)
df_copy['Price']
```
